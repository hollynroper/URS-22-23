```python
import matplotlib.pyplot as plt
import numpy as np

snr = [-12, -10, -5, 1, 5, 10]
```

```python
# values after 5 iterations, corresponding to snr values
ist = [0.02906729, 0.0267219, 0.02516862, 0.02494599, 0.02565515, 0.02452523]
amp = [0.04892012, 0.03433974, 0.02707153, 0.01524438, 0.01907732, 0.0155631]
lista = [0.022318977862596512, 0.018563086166977882, 0.010864402167499065, 0.0066844928078353405, 0.00582664180546999, 0.005395969841629267]
```

```python
plt.figure()
plt.title("MSE vs. SNR, 5 iterations")
plt.semilogy(snr, ist, label="IST", color="green")
plt.semilogy(snr, amp, label="AMP", color="orange")
plt.semilogy(snr, lista, label="LISTA", color="blue")
plt.legend()
plt.ylabel("MSE")
plt.xlabel("SNR (dB)")
plt.grid(True, which='both')
plt.show()
```
