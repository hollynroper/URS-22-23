```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_alpha = [.028, .025, .023, .021, .0196]
ist = [.0325, .0278, .0269, .0258, .025]
amp = [.0325, .0253, .0217, .0176, .016]
train_qw = [.0153, .0092, .0064, .0048, .0035]
layers = [1,2,3,4,5]
```

```python
plt.figure()
plt.plot(layers, ist, label="IST")
plt.plot(layers, amp, color="orange", label="AMP")
plt.plot(layers, train_alpha, color="red", label="Train alpha")
plt.plot(layers, train_qw, color="blue", label="Train Q, W")
plt.xlabel("Iteration Number")
plt.ylabel("MSE")
plt.grid(True, which='both')
plt.legend()
plt.show()
```
